```python
from datetime import date, timedelta
import pandas as pd
import numpy as np
```

```python
prices = {'Basic Cupcake': 5 , 'Delux Cupcake': 6 , 'Total': 1}
today = date.today()
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
print(type(today))
```

```python
basic_cupcake_file = pd.read_csv('week-1/Basic.txt', header=0 , names=['qty_basic'])
delux_cupcake_file = pd.read_csv('week-1/Delux.txt', header=0 , names=['qty_delux'])
total_cupcake_file = pd.read_csv('week-1/Total.txt', header=0 , names=['qty_total'])
```

## For total cupcakes

```python
total_cupcake_file["Year"] = pd.DatetimeIndex(total_cupcake_file["Date"]).year

total_cupcake_file["Month"] = total_cupcake_file['Date'].dt.month

total_cupcakes = total_cupcake_file.pivot_table(index="Year", columns="Month", values="qty_total", aggfunc=np.sum, fill_value=0)

total_cupcakes.set_axis(month_names,axis=1,inplace=True)

yearly_total_cupcakes_sum = list(pd.Series(total_cupcakes.sum(axis=1)))

total_cupcakes["TOTAL"] = yearly_total_cupcakes_sum
```

# Calculating total revenue for basic cupcakes

```python
basic_cupcake_file["Year"] = pd.DatetimeIndex(basic_cupcake_file["Date"]).year
basic_cupcake_file["Month"] = basic_cupcake_file['Date'].dt.month
basic_cupcakes = basic_cupcake_file.pivot_table(index="Year", columns="Month", values="qty_basic", aggfunc=np.sum, fill_value=0)
basic_cupcakes.set_axis(month_names,axis=1,inplace=True)
yearly_basic_cupcakes_sum = list(pd.Series(basic_cupcakes.sum(axis=1)))
basic_cupcakes["TOTAL"] = yearly_basic_cupcakes_sum
```

```python
basic_cupcakes
```

```python
basic_cupcakes_revenue = basic_cupcakes * 5
```

```python
basic_cupcakes_revenue
```

# Calculating total revenue of delux cupcakes

```python
delux_cupcake_file["Year"] = pd.DatetimeIndex(delux_cupcake_file["Date"]).year
delux_cupcake_file["Month"] = delux_cupcake_file['Date'].dt.month
delux_cupcakes = delux_cupcake_file.pivot_table(index="Year", columns="Month", values="qty_delux", aggfunc=np.sum, fill_value=0)
delux_cupcakes.set_axis(month_names,axis=1,inplace=True)
yearly_delux_cupcakes_sum = list(pd.Series(delux_cupcakes.sum(axis=1)))
delux_cupcakes["TOTAL"] = yearly_delux_cupcakes_sum
```

```python
delux_cupcakes
```

```python
delux_cupcakes_revenue = delux_cupcakes * 6
```

```python
delux_cupcakes_revenue
```

# Calculating total revenue of both cupcakes

```python
total_cupcakes_revenue = basic_cupcakes_revenue.add(delux_cupcakes_revenue)
```
